# Data Analysis Report of SapJam Topic "ChatBot"
Dataset file: 2800 articles crawled from SapJam by Python Scrapy and Splash

```python
from JamScrapy.mysql import MySQL
from utility import plt_pie
```

```python
db = MySQL()
posts = db.query_dic({
    'select': '*',
    'from': 'jam_post_spider',
    'where': 'body IS NOT NULL'
})

print(len(posts))
```

## Overview Jam Categories of Topic "ChatBot"

```python
urls = []
for p in posts:
    urls.append(str(p[1]).replace('http://jam4.sapjam.com/','').split('/')[0])
    
#print(urls)

urls_set = set(urls)
pie = {}
for item in urls_set:
    pie[item] = urls.count(item)
    
print(len(pie))    
print(pie)
```

Most posts fron 'groups': 2224

```python
pie.pop('groups')
values = []
labels = []
for key in pie:
    if int(pie[key]) > 0:
        values.append(int(pie[key]))
        labels.append(key)
        
plt_pie(values, labels)
```
